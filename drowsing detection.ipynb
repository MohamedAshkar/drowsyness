```python
cap = cv2.VideoCapture("https://dm0qx8t0i9gc9.cloudfront.net/watermarks/video/BKaAXTX/las-vegas-traffic-2719_x162o798_1080__d__accde87fe05d016e4745eebb9fbffd4f__P360.mp4")
while cap.isOpened():
    ret, frame = cap.read()
    result=model(frame)
    cv2.imshow("yolo", np.squeeze(result.render()))
    if cv2.waitKey(10) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import uuid
import os
import time
```

```python
IMAGES_PATH = os.path.join('data','images')
labels = ["awake","drowsy"]
number_imgs=20
```

```python
cap = cv2.VideoCapture(0)
for label in labels:
    print("collecting images for {}".format(label))
    time.sleep(5)
    for img_num in range (number_imgs):
        print("collecting images for {}, image number {}".format(label,img_num))
        ret,frame=cap.read()
        
        imgname =os.path.join(IMAGES_PATH,label+'.'+str(uuid.uuid1())+'.jpg')
        
        cv2.imwrite(imgname, frame)
        cv2.imshow("image_collection",frame)
     
        time.sleep(2)
        if cv2.waitKey(10) & 0xFF == ord("q"):
            break

cap.release()
cv2.destroyAllWindows()    
```
